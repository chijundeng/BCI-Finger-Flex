# 两算法对比分析

```python
import matplotlib.pyplot as plt
import numpy as np
# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 数据准备
methods = ['GBR', 'RF']
feature_methods = ['Wavelet', 'Raw']

# 平均RHO
rho = {
    'GBR': [0.488, 0.304],
    'RF': [0.425, 0.264]
}

# 平均MSE
mse = {
    'GBR': [0.018, 0.020],
    'RF': [0.025, 0.021]
}

x = np.arange(len(feature_methods))  # 特征提取方法的位置
width = 0.35  # 柱状图的宽度

fig, ax = plt.subplots(figsize=(10, 6))

# 绘制RHO
rects1 = ax.bar(x - width/2, rho['GBR'], width, label='GBR Wavelet')
rects2 = ax.bar(x + width/2, rho['RF'], width, label='RF Wavelet')

# 添加标签和标题
ax.set_ylabel('Average RHO')
ax.set_title('平均相关系数比较')
ax.set_xticks(x)
ax.set_xticklabels(feature_methods)
ax.legend()

# 标注
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 上移3点
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()

# 同理可以绘制MSE比较
fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x - width/2, mse['GBR'], width, label='GBR Wavelet')
rects2 = ax.bar(x + width/2, mse['RF'], width, label='RF Wavelet')

ax.set_ylabel('Average MSE')
ax.set_title('平均均方误差比较')
ax.set_xticks(x)
ax.set_xticklabels(feature_methods)
ax.legend()

autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
# 受试者1数据
subjects = ['GBR Wavelet', 'GBR Raw', 'RF Wavelet', 'RF Raw']
rho_values = [0.488, 0.304, 0.292, 0.236]
mse_values = [0.018, 0.020, 0.021, 0.020]

x = np.arange(len(subjects))
width = 0.35

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(x - width/2, rho_values, width, label='RHO', color='skyblue')
ax1.set_ylabel('RHO', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticks(x)
ax1.set_xticklabels(subjects, rotation=45)

ax2 = ax1.twinx()
ax2.bar(x + width/2, mse_values, width, label='MSE', color='salmon')
ax2.set_ylabel('MSE', color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.title('受试者1不同方法下的RHO与MSE比较')
plt.show()
```

```python
# 定义方法和特征提取方式
methods = ['GBR', 'RF']
feature_methods = ['Wavelet', 'Raw']

# 定义RHO和MSE数据
rho = {
    'GBR': [0.488, 0.304],
    'RF': [0.425, 0.264]
}

mse = {
    'GBR': [0.018, 0.020],
    'RF': [0.025, 0.021]
}

x = np.arange(len(feature_methods))  # X轴位置
width = 0.35  # 柱状图宽度

# 绘制RHO比较图
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, rho['GBR'], width, label='GBR Wavelet', color='steelblue')
rects2 = ax.bar(x + width/2, rho['RF'], width, label='RF Wavelet', color='orange')

# 添加标签和标题
ax.set_ylabel('Average RHO')
ax.set_title('GradientBoostingRegressor vs RandomForestRegressor 平均相关系数比较')
ax.set_xticks(x)
ax.set_xticklabels(feature_methods)
ax.legend()

# 添加数值标签
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 在柱子上方
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()

# 绘制MSE比较图
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, mse['GBR'], width, label='GBR Wavelet', color='steelblue')
rects2 = ax.bar(x + width/2, mse['RF'], width, label='RF Wavelet', color='orange')

ax.set_ylabel('Average MSE')
ax.set_title('GradientBoostingRegressor vs RandomForestRegressor 平均均方误差比较')
ax.set_xticks(x)
ax.set_xticklabels(feature_methods)
ax.legend()

autolabel(rects1)
autolabel(rects2)

plt.show()
```
